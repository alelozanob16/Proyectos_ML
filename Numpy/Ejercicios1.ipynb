```python
import numpy as np
import numpy.ma as ma
```

```python
#Ejercicio 1:

#En primer lugar hay que crear un array con numpy que sea del 1 al 100 pero solo los impares
array_impares = np.arange(1,100,2)
print(array_impares)

#Hay que crear un array tipo cadena que contenga nombre y apellidos
array_nombre = np.array([['Alejandro'], ['Lozano'], ['Bernal']], dtype = str)
array_nombre = np.asmatrix(array_nombre)
print(array_nombre)

#Hay que crear un array de fechas desde el 1 de enero de 2020 hasta el 31 de diciembre de 2020
array_fechas = np.arange('2020-01', '2021-01', dtype = 'datetime64[D]') 
print("Estos son los dias del año 2020: " + ", ".join(map(str, array_fechas)))
```

```python
#Ejercicio 2:

#Hay que recoger los últimos diez numeros impares del array de los 100 primeros impares
print(len(array_impares))
array_impares[-10:50]
```

```python
#Ejercicio 3:

#Hay que recorrer el array de fechas y crear uno nuevo con los dias laborables
array_dias_laborables = np.array([], dtype='datetime64[D]')
for dia in array_fechas:
    if(np.is_busday(dia)):
        array_dias_laborables = np.append(array_dias_laborables, dia)

array_dias_laborables

#Otra manera sería con el indexado booleano
array_fechas[np.is_busday(array_fechas)]
```

```python
#Ejercicio 4:

#Hay que crear la siguiente matriz:
from sys import getsizeof
matriz_E4 = np.matrix([[10,1,8,4], [3,7,2,1], [0, 2, 20, 12]], dtype='int64') #En este caso se ha creado una matriz de 3x4 de tipo int64(entero de 64 bits máximo)
matriz_E4
#Ahor vamos a generar un array en base al anterior aplicandole una máscara a los valores mayores de 10
array_enmascarado = ma.masked_where(matriz_E4 >= 10, matriz_E4)
array_enmascarado
```

```python
#Ejercicicio 5:

# En este ejercicio tenemos que generar dos fechas y calcular la diferencia entre ellas en horas y
# ver en qué fecha se convertiría si le restamos 236 horas a una de esas fechas

fecha1 = np.datetime64('1993-11-29T16:00:00')
fecha2 = np.datetime64('1998-10-06T08:00:00')
diferencia = fecha2 - fecha1
diferencia_horas = np.timedelta64(diferencia, 'h')
print(diferencia_horas)

nueva_fecha2 = fecha2 - np.timedelta64(236, 'h')
print(nueva_fecha2)# Hubiera nacido el 26 de septiembre de 1998 a las 4 de la mañana
```
