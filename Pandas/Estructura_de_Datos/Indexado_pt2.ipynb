# Indexado y Selección de Datos pt2

```python
import numpy as np
import pandas as pd
from pathlib import Path
data_path = Path('Datos_Pruebas/prueba.csv')
dataset = pd.read_csv(data_path, sep=',')
dataset
```

```python
#El método sample nos devuelve una parte, de la cual indicamos el tamaño, del dataset. Los registros que nos devuelve son aleatorios
#Este método puede llegar a ser muy útil para definir variables en pos de dividir el dataset en entrenamiento y validación
r = dataset.sample(frac=0.25)#Nos devuelve el 25% del dataset original en valores aleatorios.
r

#(Un inciso, el método shape nos devuelve las filas y las columnas del dataset)
```

```python
#La funcion .isin nos dan los registros dentro de un intervalo consecutivo.
dataset[dataset['age'].isin([20,21])]
```

```python
#También tenemos el método where. Este método nos da los registros que cumplen la condición 
# y nos da la posibilidad de sustituir los registros que no cumplen dicha condición por 
# el valor que le indiquemos.
dataset['age'].where(dataset['age'] == 20, 0)
```

```python
# Hay otro método que nos ayuda a filtrar de manera más rápida:
print(dataset.query('age == 27'), "\n")
print(dataset.query('age == 20 and genre == "HipHop"'), "\n")#Es importante destacar que la cadena va entrecomillada aunque ya este dentro
# de otra cadena
print(dataset.query('index < 5'))
```

```python
#Query también nos permite agrupar por un intervalo consecutivo:
dataset.query('age in (20,21)')
```

```python
#Si tenemos una variable con el valor por el cual queremos filtrar,
#podemos usar también query de esta manera:
edad_buscada = 20
dataset.query(f'age == {edad_buscada}')
genre_buscado = 'HipHop'
dataset.query(f'genre == "{genre_buscado}"')# Al ser un String va entrecomillado
```

```python
# Existe un método en pandas que también es muy util que es el 
# drop_duplicates:
dataset.drop_duplicates('genre', inplace=False, keep = 'first')#El keep hace que mantengamos los primeros o últimos valores
#De esta manera se pueden observar los valores únicos
```
