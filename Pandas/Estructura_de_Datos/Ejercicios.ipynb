# Una serie de ejercicios para practicar las lecciones de estructuras de datos en Pandas

```python
import numpy as np
import pandas as pd
from pathlib import Path
datapath  = Path('house.csv')
```

```python
#Ejercicio 1. En este caso vamos a resolver los distintos apartados del ejercicio 1 (tiene 3) para ello vamos a cargar el dataset house.csv
dataframe = pd.read_csv(datapath, sep = ',')
dataframe.head()
```

```python
#En el primer apartado del ejercicio 1, se nos pide que busquemos si existe alguna casa con 8 habitaciones:
casa8 = dataframe[dataframe['bedrooms'] == 8]
casa8
```

```python
#En el segundo apartado del ejercicio 1, nos pide el mínimo y el máximo de habitaciones que tiene una casa:
minimo = dataframe['bedrooms'].min()
maximo = dataframe['bedrooms'].max()
print(minimo)
print(maximo)
```

```python
#En el tercer apartado del ejercicio 1, nos pide que creemos una nueva columna con el precio por habitación de la casa:
dataframe['price_by_floor'] = dataframe['price'] / dataframe['floors'] #Asi creamos una nueva columna con el precio por habitación
dataframe.drop(columns='prize_by_floor', inplace=True) #Borramos esta columna dado que su nombre esta en español
dataframe.head()
```

```python
#Ejercicio 2. Tiene 4 apartados que vamos a ir resolviendo.

#El primer apartado nos dice que le digamos el precio de la casa 256:
print(dataframe.loc[256, 'price'])
```

```python
#Nos dice que le digamos el número de habitaciones de las casas de 215 a la 222:
habitaciones = dataframe.loc[215:222, 'bedrooms']
habitaciones
```

```python
#Tenemos que recuperar un sample aleatorio del 15% de las casas:
sample15 = dataframe.sample(frac=0.15)
sample15
#Una vez tengamos el sample del 15%, debemos filtrar las casas con 3 o 4 habitaciones cuyo precio sea mayor a 300_000€

habitaciones_3_4_300000 = sample15[sample15['bedrooms'].isin([3, 4]) & (sample15['price'] > 300_000)]
habsss = sample15.query('bedrooms in [3, 4] and price > 300000') #Otra opción de hacerlo con .query
print(habsss['price'].min())
print(habitaciones_3_4_300000['price'].min())
print(habsss['price'].max())
print(habitaciones_3_4_300000['price'].max())
#De esta manera comprobamos que son los mismos resultados.
```

```python
#El ejercicio 3 nos pide que creemos un multiindex que agrupe una serie de columnas que ahora veremos bajo localizacion 
#y el resto bajo caracteristicas:
#Vamos a crear el multiindex en base a tuplas:
dataframe.columns
tuples = [('Caracteristicas', 'date'),
          ('Caracteristicas', 'price'),
          ('Caracteristicas', 'bedrooms'),
          ('Caracteristicas', 'bathrooms'),
          ('Caracteristicas', 'sqft_living'),
          ('Caracteristicas', 'sqft_lot'),
          ('Caracteristicas', 'floors'),
          ('Caracteristicas', 'waterfront'),
          ('Caracteristicas', 'view'),
          ('Caracteristicas', 'condition'),
          ('Caracteristicas', 'sqft_above'),
          ('Caracteristicas', 'sqft_basement'),
          ('Caracteristicas', 'yr_built'),
          ('Caracteristicas', 'yr_renovated'),
          ('Caracteristicas', 'price_by_floor'),
          ('Localización', 'street'),
          ('Localización', 'city'),
          ('Localización', 'statezip'),
          ('Localización', 'country'),]
multiindex = pd.MultiIndex.from_tuples(tuples)
dataframe.columns = multiindex
dataframe.head()
df_location = dataframe['Localización']
df_location.head(20)
```

```python
#El ejercicio 4 tiene otros 4 apartados:

#En el primero se nos pide que convirtamos el campo de date en un tipo fecha(date):
dataframe['Caracteristicas', 'date'] = pd.to_datetime(dataframe['Caracteristicas', 'date'], yearfirst=True)
dataframe['Caracteristicas', 'date']
```

```python
#En el segundo:
#Se nos pide que creemos una columna con el año, otra con el mes y otra con el día:
dataframe['anyo'] = dataframe['Caracteristicas', 'date'].dt.year
dataframe['mes'] = dataframe['Caracteristicas', 'date'].dt.month
dataframe['dia'] = dataframe['Caracteristicas', 'date'].dt.day
dataframe.head()
```

```python
#En el tercero nos pide que hagamos el precio medio por mes:
#Para ello vamos a usar el resample, pero primero tenemos que tener la fecha como indice:
data = pd.read_csv(datapath, sep = ',')
data['date'] = pd.to_datetime(data['date'], yearfirst=True)
data.set_index('date', inplace=False).resample('1MS')['price'].mean()
```

```python
#El ultimo punto nos pide que creemos una nueva columna fruto de sumar 20 dias a la columna fecha:
td_20 = pd.Timedelta(20, unit='D')
data['date+20'] = data['date'] + td_20
data.head()
```
