# Vamos a ver operaciones entre dataframes con Pandas

```python
import pandas as pd
import numpy as np
from pathlib import Path
data_path = Path('../Estructura_de_Datos/house.csv')
```

```python
data = pd.read_csv(data_path, sep=',')
data.head()
```

```python
# Nos quedamos con las primeras 10 filas y las Ãºltimas 10
df1 = data.head(10)
df2 = data.tail(10)
```

```python
df1
```

```python
df2
```

```python
# Concatenamos ambos dataframes
df = pd.concat([df1, df2], axis = 0) #Le pasamos una lista con los dataframes a concatenar y con axis = 0 le decimos 
# que los una hacia abajo (uno debajo del otro) y con axis = 1 los unirÃ­a hacia la derecha (uno al lado del otro)
df.shape
```

```python
# TambiÃ©n podemos concatenarlos aÃ±adiendole un multiindice:
df = pd.concat([df1, df2], keys = ['first 10 rows', 'last 10 rows'])
df
```

```python
df.loc['first 10 rows'].iloc[-5:] # TambiÃ©n podemos acceder a los datos de un multiindice
```

```python
# Ahora vamos a unir dataframes con distintas columnas
df1 = data.head(10)[['price', 'bedrooms']]
df2 = data.head(10)[['city', 'statezip', 'country']]
df  = pd.concat([df1, df2], axis = 1)#Lo unimos hacia la derecha
df
```

```python
df1 = data.head(10)[['city', 'price', 'bedrooms']]
df2 = data.head(10)[['city', 'statezip', 'country']]
df1
```

```python
df2
```

```python
# Vamos a utilizar el merge para unir ambos dataframes
df = df1.merge(df2, how = 'inner', on = 'city')
df
#El metodo merge nos permite unir dos dataframes a partir de una columna en comÃºn, en este caso 'city'.
#El argumento how = 'inner' nos indica que solo se queden las filas que tengan un valor en comÃºn en la columna 'city'
#El argumento on = 'city' nos indica que la columna en comÃºn es 'city'
#Si fuese un left join, se quedarÃ­an todas las filas de df1 y solo las filas de df2 que tengan un valor en comÃºn en la columna 'city'
#Si fuese un right join, se quedarÃ­an todas las filas de df2 y solo las filas de df1 que tengan un valor en comÃºn en la columna 'city'
#Si fuese un outer join, se quedarÃ­an todas las filas de df1 y df2 que tengan un valor en comÃºn en la columna 'city'
#En este caso como todos los valores de city son iguales en ambos dataframes, 
#el resultado es el mismo que con un inner join que con un outer join o con un left join o con un right join
```
