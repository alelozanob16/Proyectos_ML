# Vamos a realizar una serie de operaciones básicas con un dataframe. Los datos que usaremos serán los de los ejercicios anteriores

```python
import pandas as pd
import numpy as np
from pathlib import Path
datapath = Path('../Estructura_de_datos/house.csv')
```

```python
data = pd.read_csv(datapath, sep=',')
data.head()
print(data.shape)#Sabemos el número de filas y columnas
print(data.columns)#Sabemos el nombre de las columnas
print(data.dtypes)#Sabemos el tipo de dato de cada columna
```

```python
data.describe()#Nos da un resumen de los datos
```

```python
data.mode()#Nos da la moda de cada columna
data.min()#Nos da el valor mínimo de cada columna
data.max()#Nos da el valor máximo de cada columna
data.sum()#Nos da la suma de cada columna
```

```python
data.isna().sum()#Nos dice cuantos valores nulos hay en cada columna
```

```python
data['bedrooms'].value_counts(dropna=False, normalize=True)#Nos dice cuantas veces se repite cada valor en la columna price, 
# con dropna=False incluye los valores nulos y normalize nos va a normalizar los valores
```

```python
#Otros metodos interesantes son el qcut y el cut. En el caso del qcut nos va a dividir los datos en n grupos de igual tamaño
#En el caso del cut nos va a dividir los datos en n grupos de igual rango
pd.qcut(data['price'], q=4).value_counts()#Nos divide los datos en 4 grupos de igual tamaño
```

```python
pd.cut(data['bedrooms'], bins=2).value_counts()#Nos divide los datos en 2 grupos de igual rango, pero distinto tamaño
#Podemos ponerle los intervalos que nos de la gana:
pd.cut(data['bedrooms'], bins=[0, 3, 8, 10], labels=['menos de 3', 'entre 3 y 8', 'entre 8 y 10'])
```

```python
data.sort_values(by='bedrooms', ascending=True, inplace=False)#Nos ordena los datos en función de la columna price de menor a mayor 
```

```python
#El metodo copy nos permite hacer una copia de un dataframe
data2 = data.query('bedrooms == 0')
data2
```

```python
d = data2.copy(deep=True)
d
```

```python
data2.iat[0,2] = 10
data2
```

```python
d #Cuando el parametro de deep = False, si modificamos el dataframe original, también se modifica el dataframe copia
d #Cuando el parametro de deep = True, si modificamos el dataframe original, no se modifica el dataframe copia
```
